```python
#Merge Data
df = data["olist_orders_dataset"].merge(data["olist_customers_dataset"], on="customer_id", how="left")
order_items_products = data["olist_order_items_dataset"].merge(data["olist_products_dataset"], on="product_id", how="left")
df = df.merge(order_items_products, on="order_id", how="left")
df = df.merge(data["olist_order_payments_dataset"], on="order_id", how="left")
df = df.merge(data["olist_order_reviews_dataset"], on="order_id", how="left")
df = df.merge(data["olist_sellers_dataset"], on="seller_id", how="left")
```

```python
#AOV & Retention
df['total_price'] = df['price'] * df['order_item_id']
```

```python
aov = df.groupby('order_id')['total_price'].sum().mean()
print("Average Order Value:", round(aov,2))
```

```python
repeat_orders = df.groupby('customer_unique_id')['order_id'].count()
retention_rate = (repeat_orders > 1).mean() * 100
print("Customer Retention Rate:", round(retention_rate,2), "%")
```

```python
#Sentiment Analysis
from textblob import TextBlob
```

```python
df['review_sentiment'] = df['review_comment_message'].apply(lambda x: TextBlob(str(x)).sentiment.polarity)
df['sentiment_type'] = df['review_sentiment'].apply(lambda x: 'Positive' if x>0 else ('Negative' if x<0 else 'Neutral'))

```

```python
sns.countplot(data=df, x='sentiment_type')
plt.title("Review Sentiment Distribution")
plt.show()
```

```python
#Delivery Analysis
df['delivery_delay'] = (df['order_delivered_customer_date'] - df['order_estimated_delivery_date']).dt.days
```

```python
sns.histplot(df['delivery_delay'].dropna(), bins=30)
plt.title("Delivery Delay Distribution")
plt.xlabel("Days")
plt.ylabel("Number of Orders")
plt.show()
```

```python
#Simple Regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
X = df.groupby('order_id')['order_item_id'].count().reset_index().rename(columns={'order_item_id':'num_items'})
y = df.groupby('order_id')['total_price'].sum().reset_index()['total_price']
X_train, X_test, y_train, y_test = train_test_split(X[['num_items']], y, test_size=0.2, random_state=42)

```

```python
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("RMSE:", mean_squared_error(y_test, y_pred, squared=False))
```

```python
#Time-Series Trend
df['order_month'] = df['order_purchase_timestamp'].dt.to_period('M')
monthly_sales = df.groupby('order_month')['total_price'].sum()
monthly_sales.plot(kind='line')
plt.title("Monthly Sales Trend")
plt.ylabel("Total Revenue")
plt.show()
```

```python
from dash import Dash, html, dcc
import pandas as pd
import plotly.express as px

# Line chart data (Orders over time)
orders_df = pd.DataFrame({
    "month": ["Jan", "Feb", "Mar", "Apr"],
    "orders": [1200, 1500, 1800, 2100]
})

# Bar chart data (Products by category)
products_df = pd.DataFrame({
    "category": ["Electronics", "Furniture", "Fashion"],
    "count": [300, 180, 520]
})

# Pie chart data (Sentiment distribution)
sentiment_df = pd.DataFrame({
    "sentiment": ["Positive", "Neutral", "Negative"],
    "reviews": [800, 250, 150]
})

fig_orders = px.line(
    orders_df,
    x="month",
    y="orders",
    title="Monthly Orders Trend"
)

fig_products = px.bar(
    products_df,
    x="category",
    y="count",
    title="Products by Category"
)

fig_sentiment = px.pie(
    sentiment_df,
    names="sentiment",
    values="reviews",
    title="Customer Review Sentiment"
)

#Dashboard
app = Dash(__name__)

app.layout = html.Div([
    html.H1("Olist E-Commerce Dashboard", style={'textAlign': 'center'}),

    dcc.Graph(figure=fig_orders),
    dcc.Graph(figure=fig_products),
    dcc.Graph(figure=fig_sentiment)
])


if __name__ == "__main__":
    app.run(debug=True)
```
